```python
from IPython.display import display
import os
import logging


import matplotlib.pyplot as plt
import numpy as np
import simpleaudio as sa

fs = 44100 # sample rate
channels = 1

def play_sound(data, channels):
    c = np.empty((data.size,), dtype=data.dtype)
    length = int(len(data)/channels)
    for i in range(channels):
        c[i::channels] = data[i*length:(i+1)* length]
    play_obj = sa.play_buffer(c, channels, 2, fs)
    play_obj.wait_done()

FOLDER = '/home/alex/.wine/drive_c/GOG Games/WH40K Final Liberation/MUSIC'

for f in sorted(os.listdir(FOLDER))[:1]:
    if '.mus' in f:
        with open(FOLDER + '/' + f, 'rb') as file:
            data = np.fromfile(file, dtype=np.int16)
        print(f, "{} m".format(round(len(data) / fs / 60 / channels, 2)))
        print(len(data))
#         plt.plot(np.linspace(0, 1, len(data) - 12), data[12:])
        play_sound(data[12:], channels)
#         plt.show()
```
